```python
# import libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
import collections
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

# Data Cleaning

The original dataset wasn't that bad and only needed a little of cleaning. We first removed the columns that we didn't need and the identification to maintain
ethicacy. The biggest problem was the opening_eco and opening variations, because there are multiple types of variation that you can do. We first wanted to 
scrape a website containing the ecos and store it into a dictionary with a standardized format where all variations of an opening will effectively be the same one.
Next we replaced the opening_ecos and removed the opening_name because they contained variations, which we didn't want. We then added a new column for rating difference.


```python
# open data set
games = pd.read_csv('games.csv')
```

```python
games.head()
```

```python
# check if data is complete

if games.isnull().sum().sum() == 0:
    print('no NaN')
else:
    print('Nan detected')
```

```python
# remove unwanted columns

games_cols = []

for col in games.columns:
    games_cols.append(col)
    
print(games_cols)
print(len(games_cols))
    
wanted_cols = ['winner',
               'white_rating',
               'black_rating',
               'opening_eco'
              ]
# we only want the winners, ratings, and openings
games = games[wanted_cols]
```

```python
games.head()
```

```python
## Scrape openings from https://www.365chess.com/eco.php so we can edit it onto our data set and get desired openings

# get ecos
eco_url = 'https://www.365chess.com/eco.php'
page = requests.get(eco_url)

# parse eco page
soup = BeautifulSoup(page.content, "html.parser")

# scrape ecos
x = soup.find_all(class_ = "opname")


# get raw ecos html
opnames = []
for y in x:
    opnames.append(y.text)
    

# parse raw ecos html into dict
opening_dict = {}

for i in opnames:
    split = i.split()
    if split[0] not in opening_dict:
        opening_dict[split[0]] = ' '.join(split[1:])

# only entry that's messed up, just manually fix it
opening_dict['D50-D69'] = "Queen's Gambit Declined"
```

```python
## clean dict of ecos and subsets of major openings 

# get range of '-'
def get_range(a):
    num_range = []
    for x in a:
        num_range.append(int(re.sub('\D', '', x)))
    return num_range
        

# create new entries based on major openings and delete old ones
numbers = []
new_dict = opening_dict.copy()

for key in opening_dict:
    if '-' in key:
        numbers = (key.split('-'))
        # num_range = 69
        num_range = get_range(numbers)
        # num_ranges = 10 - 00
        num_ranges = num_range[1] - num_range[0] + 1
        alpha_key = numbers[0][0]
        for x in range(num_ranges):
            num_str = str((num_range[0] + x))
            if len(num_str) < 2:
                num_str = alpha_key + '0' + num_str
            else:
                num_str = alpha_key + num_str
            new_dict[num_str] = opening_dict[key]
        del new_dict[key]
    


op_eco_dict = collections.OrderedDict(sorted(new_dict.items()))
```

```python
# successfully replace opening_eco
games = games.replace({'opening_eco' : op_eco_dict}) 
games = games.rename(columns = {'opening_eco': 'opening'})


# round games rating
games['white_rating'] = games['white_rating'].round(-2)
games['black_rating'] = games['black_rating'].round(-2)

# get rating diff
games['rating diff'] = abs(games['white_rating'] - games['black_rating'])


# split by winner color now
black_win = games[games['winner'] == 'black'].reset_index(drop = True)
white_win = games[games['winner'] == 'white'].reset_index(drop = True)
```

```python
# Get sub data sets on different elo groups
'''
- **Beginner** : below 1200
- **Intermediate** : 1200-1800
- **Advanced** : 1800 - 2000
- **Expert** : above 2000
'''
black_beg_win = black_win[black_win['black_rating'] <= 1200]
white_beg_win = white_win[white_win['white_rating'] <= 1200]

black_inter_win = black_win[black_win['black_rating'].between(1200,1800)]
white_inter_win = white_win[white_win['black_rating'].between(1200,1800)]

black_adv_win = black_win[black_win['black_rating'].between(1800,2000)]
white_adv_win = white_win[white_win['black_rating'].between(1800,2000)]

black_exp_win = black_win[black_win['black_rating'] >= 2000]
white_exp_win = white_win[white_win['white_rating'] >= 2000]

beg_win = pd.concat([black_beg_win,white_beg_win ])
inter_win = pd.concat([black_inter_win,white_inter_win])
adv_win = pd.concat([black_adv_win,white_adv_win])
exp_win = pd.concat([black_exp_win,white_exp_win])
```

Interestingly, the Sicilian isn't played a lot in beginner games, as beginners rather opt for a King's pawn game, which is a safer and more recommended opening for beginners, as the Sicilian has many difficult variantions and could get out of hand.

Let's examine the win rates of the top 5 opening in each skill level, and since our data set is kind of small, let's restrict it to at least 50 games played

```python
beg_op =[]
for i in range(len(beg_win['opening'].value_counts() > 50)):
    if (beg_win['opening'].value_counts() > 50)[i]:
        beg_op.append((beg_win['opening'].value_counts() > 50).index[i])
```

```python
# get all top openings for beginners
beg_op_df = []
for i in beg_op:
    beg_op_df.append(beg_win[beg_win['opening'] == i])

white_win_rate = []

for i in range(len(beg_op_df)):
        white_win_rate.append(len(beg_op_df[i][beg_op_df[i]['winner'] == 'white']) / len(beg_op_df[i]))

white_beg_win_dict = {beg_op[i]: white_win_rate[i] for i in range(len(beg_op))}

white_beg_win_df = pd.DataFrame(list(white_beg_win_dict.items()))
white_beg_win_df.columns = ['opening','win rate']
```

```python
plt.figure(figsize = (15,8))
plt.xticks(rotation=90)
sns.barplot(x = 'opening', y = 'win rate',data =white_beg_win_df)
```

As we can see, our theory is true where the king's pawn opening may be better than the sicilian, as the king's pawn is more fundamental and is easier to play. What's also interesting is that the variation of the king's pawn opening, the king's pawn game, where we move a knight after our pawn opening has a low win percentage, which is weird, but the other variation, the king's knight opening, has an extremely high win rate. Nevertheless, we can see that the king's pawn opening has an extremely high win rate, as long as you move e4.
